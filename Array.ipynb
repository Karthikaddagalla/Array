### The below class(array1) can be used to create  arrays in python similar to arrays in  "C" 

```python
#Original code

class Array1:
    def __init__(self, typecode,grades):
        self.typecode = typecode
        self.grades = grades
        global origin
        origin = {"i": int, "f": float, "I":int, "u": str, "b": int, "B": int,  "h": int, "H": int, "l": int, "L": int, "q": int, "Q": int, "d": float}
        for i in self.grades:
            if origin[typecode] != type(i):
                raise TypeError(f"We are expecting a {origin[typecode]}, but {type(i)} is given")
                
                
    def __len__(self):
        count = 0
        for i in self.grades:
            count+=1
        return count
    
    
    def __getitem__(self,i):
        return self.grades[i]
    
    
    def __repr__(self):
        m = ""
        for i in self.grades:
            m = m+str(i)
        return f"array('{self.typecode}', \'{m}' )"
    
    
    def index(self,val):
        for i in range(0, len(self.grades)):
            if self.grades[i] == val:
                return i
        else:
            raise ValueError(f"{val} is not found in array")
            
            
    def append(self, k):
        if type(k) is not origin[self.typecode]:
             raise TypeError("array item must be unicode character")
                
        self.grades = self.grades + [k]
            
    
    def extend(self,it):
        for i in it:
            if type(i) is not origin[self.typecode]:
                raise TypeError(f"We are expecting a {origin[self.typecode]}, but {type(k)} is given")
            self.append(i)
            
            
    def pop(self, val = "a"):
        if val != "a":
            m = self.grades[val]
            del self.grades[val]
            return m
        val = self.grades[len(self.grades) - 1]
        del self.grades[len(self.grades) - 1]
        return val
            
            
    def insert(self,pos,val):
        origin = {"i": int, "f": float, "I":int, "u": str}
        if type(pos) is not int:
                 raise TypeError(f"We are expecting a {int}, but {type(pos)} is given")
        elif type(val) is not origin[self.typecode]:
                 raise TypeError(f"We are expecting a {origin[self.typecode]}, but {type(val)} is given")
        
        elif len(self.grades) == pos:
            self.grades.append(val)
            
        else:
            for i in range(len(self.grades), pos, -1 ):
                if i == len(self.grades):
                    self.grades.append(self.grades[i-1])
                else:
                    self.grades[i] = self.grades[i-1]    
            self.grades[pos] = val
    
    
    def remove(self,val):
        if type(val) != origin[typecode]:
                 raise TypeError(f"We are expecting a {origin[typecode]}, but {type(val)} is given")
        for i in range(self.index(val), len(self.grades) ):
            if i == len(self.grades) - 1:
                del self.grades[len(self.grades) - 1]
            else:
                self.grades[i] = self.grades[i+1]
    
    
    def reverse(self):
        for i in range(0, len(self.grades) // 2):
            k = self.grades[i]
            self.grades[i] = self.grades[len(self.grades)-i-1]
            self.grades[len(self.grades)-i-1] = k
    
    
    def count(self,x):
        count = 0
        for i in self:
            if i == x:
                count = count + 1
        return count
    
```
